## What is deep learning?

Machine learning is turning things (data) into numbers **finding patterns** in those numbers. 

**How to find those patterns?**  By useing code (computer algorithms) and math.

**Machine Learning vs. Deep Learning**

"Deep Learning" is a subset of "Machine Learning", and it is a subset of "Artificial Intelligence"

**Traditional programing vs Machine learning algorithm**

In traditional programming you give inputs and rules program makes an output, in machine learning algorithm you give input and and some desired output and figures out the pattern between them, figures out the rules. (Supervised Learning)

**Why use machine learning (or deep learning)?**

We as humans can't think all the rules for complex problems.

**What deep learning is good for?**

- *Problems with long lists of rules*-when the traditional approach fails, machine learning/deep learning may help.

- *Continually changeing environments*-deep learning can adapt('learn') to new senerios.

- *Discovering insights within large collections of data*

**What deep learning is not good for?**

- *When you need expainability*-the pattern learned by a deep learning model are typically uninterpretable by a human.

- *When the traditional approach is a better option*-if you can accomplish what you need with simple rule-based system. 

- *When errors are unaccaptable*- since the outputs of deep learning aren't always predictible.

- *When you don't have much data*-deep learning models usually require a fairly large amount of data to produce great results.
